### Styling in Seaborn

```python
#using grid for styling

sns.set_style(style='darkgrid')
sns.scatterplot(x='rate', y='approx_cost(for two people)', data=files, hue='rate')
```

```python
#Customizing Bin for Histogram

sns.histplot(x=files['rate'])
```

```python
#Changing the size of a plot

files['rest_type']
```

```python
sns.countplot(x=files['rest_type'])
```

```python
#To change the size of the plot

sns.set(rc={'figure.figsize':(22,12)})
sns.histplot(x=files['rest_type'])
```

```python
## Now, rotating the axis label 

sns.set(rc={'figure.figsize':(20,12)})

axes = sns.histplot(x=files['rest_type'])
axes.set_xticklabels(axes.get_xticklabels(),
                     rotation=90)
```

### How to set limit on X and Y axis

```python
##Let use scatter plot for this.. for Xlimit
sns.set(rc={'figure.figsize':(20,18)})

axes = sns.scatterplot(x='rate', y='approx_cost(for two people)', data=files)
axes.set(xlim=(4.5, 5.0))
```

```python
sns.set(rc={'figure.figsize':(20,18)})

axes = sns.scatterplot(x='rate', y='approx_cost(for two people)', data=files)
axes.set(xlim=(4.5, 5.0))
axes.set(ylim=(1500,3000))
```

### Facet-Grid in Seaborn

```python
#We first find out the total unique or variables in the column

print(files['location'].unique())
print(files['online_order'].unique())
```

```python
##Lets start with the code

ax = sns.FacetGrid(files, col='online_order',     #This show the grid 
                  row='location')                #without plots
ax.map_dataframe(sns.histplot, x='rate')          #Here, we plot
```
