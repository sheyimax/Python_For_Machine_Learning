```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.sin(90)                #sin function
b = np.cos(90)                #cos function
c = np.tan(90)                #tan function

print(f'sin(90) = {a}')        #the result is always in radians
b = np.cos(90)
print(f'cos(90) = {b}')
print(f'tan(90) = {c}')
```

```python
# Using matplotlib to plot numpy data

sin_x = np.arange(0, 4*np.pi, .2)
sin_y = np.sin(sin_x)

print(sin_x)
print(sin_y)
```

```python
#Using the matplotlib function to plot the data

plt.plot(sin_x, sin_y, 'r')              #this means plt.plot(x_axis, y_axis)
plt.show()
```

```python
cos_x = np.arange(0, 10, .2)
cos_y = np.cos(cos_x)

print(cos_x)
print(cos_y)
```

```python
plt.plot(cos_x, cos_y)
plt.show()
```

```python
tan_x = np.arange(0, 20, .2)
tan_y = np.tan(tan_x)

print(tan_x)
print(tan_y)
```

```python
plt.plot(tan_x, tan_y)
plt.show()
```

```python
#Lets define another code sample for nimpy and matplotlib

n = 100
x = np.linspace(0, 4, n+1)
y = np.exp(-x)*np.sin(2*np.pi*x)

print(x)
print(y)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import random
```

```python
# 1D array function call for random

i = np.random.random(5)                  #this gives 5 elements between range 0-1
j = np.random.random(10)                   #this gives 10 elements

print(i)
print(j)
```

```python
# 2D array function call

l = np.random.random((4,3))
m = np.random.random((4,4))

print(l)
print(m)
```

```python
# To get an integer in 1D array with specific range

#np.random.randint(random_start,random_end)
n = np.random.randint(1,6)      #This gives an int from 1-6
o = np.random.randint(9,18)      #This gives an int from 9-18

print(n)
print(o)
```

```python
# To get an integer in 2D array with specific range

#np.random.randint(random_start,random_end, (row, column))
p = np.random.randint(1,7,(4,4))
q = np.random.randint(9,15,(3,3))

print(p)
print(q)
```

```python
# To get an integer in 3D array with specific range
#np.random.randint(random_start,random_end(dimension,row,column))

r = np.random.randint(1,7,(2,4,4))
s = np.random.randint(4,9,(3,3,3))

print(r)
print(s)
```

```python
# Using random.seed() function to get the same values

t = np.random.seed(1)
t = np.random.randint(1,7,(2,4,4))
u = np.random.seed(2)
u = np.random.randint(1,7,(2,4,4))

print(t)
print(u)
```

```python
#seed value you can assigned will be

seed_value = 2**32-1

print(f'seed value can be assigned from 1 to {seed_value}')
```

```python
#For Negative random values

v = np.random.randn(4,4)

print(v)
```

```python
#Using random.choice() to select out of your choice function

list_1 = [1, 6, 9, 10, -5, -3]
w = np.random.choice(list_1)

print(w)
```

```python
#Using random.permutation() to select and combine a list in different combination

h = np.random.permutation(list_1)

print(h)
```
