### Lets visualize the data

```python
fig = plt.figure(figsize=(6,6))     #set up the figure in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, 
                    hspace=0.05, wspace=0.05)

#Lets plot the digits: remember image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary,
             interpolation='gaussian')
    
    #label the image with the target value
    ax.text(0, 7, str(digits.target[i]), style='italic', color='red',
           fontsize=8)

    
plt.show()
```

### Now lets create a scatterplot of the features ashand color intensity of the wine data set.

```python
figure, ax=plt.subplots()

features = 'ash', 'color_intensity'

features_index = [wine.feature_names.index(features[0]),
                 wine.feature_names.index(features[1])]

colors = ['red', 'blue', 'green']

for label, color in zip(range(len(wine.target_names)), colors):
    ax.scatter(wine.data[wine.target==label, features_index[0]],
               wine.data[wine.target==label, features_index[1]],
               label=wine.target_names[label],
               c=color)
    
ax.set_xlabel(wine.feature_names[wine.feature_names.index(features[0])])
ax.set_ylabel(wine.feature_names[wine.feature_names.index(features[1])])
ax.legend(loc='best')
plt.show()
```

```python
#when creating a scatterplot we have to change the datasets to a dataframe

wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)

wine_df.head()
```

```python
#lets plot the graph
def rotate_labels(df, axes):
    """changing the rotation of the label output,
    y labels horizontally and x labels vertically"""
    n = len(df.columns)
    for x in range(n):
        for y in range(n):
            ax= axs[x, y]                     #to get the axis of subplots
            ax.xaxis.label.set_rotation(90)   #to make x axis name vertical
            ax.yaxis.label.set_rotation(0)    #to make y axis name horizontal
            ax.yaxis.labelpad = 50            #to make sure y axis names are outside the plot area

axs = pd.plotting.scatter_matrix(wine_df, c=wine.target, figsize=(10,10))

rotate_labels(wine_df, axs)
plt.show()
```

### Fetch the Olivetti faces dataset and visualize the faces.

```python
#we need to fetch the dataset first

olivetti = datasets.fetch_olivetti_faces()

olivetti.keys()             #viewing the attributes of the dataset
```

```python
#lets view the description

print(olivetti.DESCR)
```

```python
#lets check the data shape

olivetti.data.shape
```

#### now write the code to visualize the faces

```python
#lets set up the figure

fig = plt.figure(figsize=(6,6))

fig.subplots_adjust(left=0, right=1, bottom=0, top=1.05,
                   hspace=0.05, wspace=0.05)

#Now, lets plot the images: we have 400 faces
for i in range(144):
    axes = fig.add_subplot(12, 12, i+1, xticks=[], yticks=[])
    axes.imshow(olivetti.images[i], cmap=plt.cm.bone, interpolation='nearest')
    
    #lets label the images with the target value
    axes.text(0, 7, str(olivetti.target[i]), fontsize=6)

plt.show()
```
