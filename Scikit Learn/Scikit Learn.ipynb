#### For Histograms of the features

```python
#for the petal width(cm)

fig, ax = plt.subplots()
x_index = 3
colors = ['red', 'blue', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index],
           label=iris.target_names[label],
           color=color)
    
ax.set_xlabel(iris.feature_names[x_index])
ax.legend(loc='best')
plt.show()
```

```python
#Lets plot for petal length and sepal width

figure, ax = plt.subplots()
x_index = 2
y_index = 1
colors = ['red', 'green', 'blue']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.scatter(iris.data[iris.target==label, x_index],
              iris.data[iris.target==label, y_index],
              label=iris.target_names[label],
              color=color)
    
ax.set_xlabel(iris.feature_names[x_index])
ax.set_ylabel(iris.feature_names[y_index])
ax.legend(loc='best')
plt.show()
```

### Generalization, Plot all features combination in one diagram

```python
n = len(iris.feature_names)
figure, ax = plt.subplots(n, n, figsize=(16,16))

colors = ['red', 'green', 'blue']

for x in range(n):
    for y in range(n):
        xname = iris.feature_names[x]
        yname = iris.feature_names[y]
        for color_ind in range(len(iris.target_names)):
            ax[x, y].scatter(iris.data[iris.target==color_ind, x],
                             iris.data[iris.target==color_ind, y],
                             label=iris.target_names[color_ind],
                             color=colors[color_ind])
            
        ax[x, y].set_xlabel(xname)
        ax[x, y].set_ylabel(yname)
        ax[x, y].legend(loc='lower right')
    

plt.show()        
```

### 3 Dimensional Visualization

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
## still using the iris dataset

X = []                                      #create an empty list
for iclass in range(3):                     #let the list be a list of 3 variables
    X.append([[], [], []])                  #we append the X to be a 3 list of lists inside
    for i in range(len(iris.data)):         #create a for loop
        if iris.target[i] == iclass:       
            X[iclass][0].append(iris.data[i][0])
            X[iclass][1].append(iris.data[i][1])
            X[iclass][2].append(sum(iris.data[i][2:]))
            
colors = ('r', 'g', 'b')
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for iclass in range(3):
    ax.scatter(X[iclass][0], X[iclass][1], X[iclass][2],
              c=colors[iclass])
    
plt.show()
```
