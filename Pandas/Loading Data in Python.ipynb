```python
# If we want to change the header to another headers of our type, 
#this can be done using the skiprows() in conjunction with names

emp = pd.read_csv('emp10.csv', skiprows=1, names=['Emp_ID','Emp_Name','Emp_Desig','Emp_DOJ','Emp_Sal'])
print(emp)
```

```python
#Using skiprows

emp = pd.read_csv('emp10.csv', skiprows=4)      #This will skip the first 4 rows including the header, i.e starts counting from 1 with the header
print(emp)
```

```python
emp = pd.read_csv('emp10.csv', skiprows=[1,2,3,4])      #This will skip the index 1,2,3,4 and leave the header
print(emp)
```

```python
#Taking header=None creates another header for you with index 0,1,2.....

emp = pd.read_csv('emp10.csv', header=None)      #This will skip the index 1,2,3,4 and leave the header
print(emp)
```

```python
#You can also create prerfixes for the header using the argument prefix=''

emp = pd.read_csv('emp10.csv', header=None, prefix='Column')      #This will skip the index 1,2,3,4 and leave the header
print(emp)
```

```python
#How to specify/know the missing or blank values

emp = pd.read_csv('emp10.csv', na_values=['.'])     
print(emp)
```

```python
#Changing column type while importing CSV

emp = pd.read_csv('emp10.csv')     
emp.dtypes
```

```python
#Now we want to change the salary to a float type

emp = pd.read_csv('emp10.csv', dtype={'sal':'float'})     
print(emp)
```

```python
#How to measure time taken to import big CSV file

emp = pd.read_csv("emp10.csv", verbose=True)
```
