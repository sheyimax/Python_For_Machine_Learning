```python
b.head(2)
```

```python
plt.scatter(x='rate', y='approx_cost(for two people)', data=b)
```

```python
#How to change color of points on scatterplot

plt.scatter(x='rate', y='approx_cost(for two people)', data=b, c='m')
```

```python
#How to change the size of the point

plt.scatter(x='rate', y='approx_cost(for two people)', 
            data=b, c='g', s=10)
```

```python
#Labelling your plot with title, x-axis and y-axis

plt.scatter(x='rate', y='approx_cost(for two people)', 
            data=b, c='g', s=10)
plt.title('A Scatter Plot for Zomata')
plt.xlabel('Rate')
plt.ylabel('Approx Cost for 2 people')
```

```python
#How to set a limit on x-axis and y-axis

plt.scatter(x='rate', y='approx_cost(for two people)', 
            data=b, c='r', s=10)
plt.title('A Scatter Plot for Restuarant Ratings')
plt.xlabel('Rate')
plt.ylabel('Approx Cost for 2 people')


#setting the limit for x-axis
plt.xlim(4.50,5.0)
plt.show()
```

```python
#How to set a limit on x-axis and y-axis

plt.scatter(x='rate', y='approx_cost(for two people)', 
            data=b, c='r', s=10)
plt.title('A Scatter Plot for Restuarant Ratings')
plt.xlabel('Rate')
plt.ylabel('Approx Cost for 2 people')


#setting the limit for x-axis
plt.xlim(4.50,5.0)
plt.ylim(1000,2000)
plt.show()
```

```python
## How to customize size of a plot
#Problem statement: What are the most famous restaurants in Bangalore

b['rest_type'].value_counts()
```

```python
#Lets get the index

b['rest_type'].value_counts().index
```

```python
#Let's plot the bar chat to represent this data

plt.bar(x=b['rest_type'].value_counts().index,
       height=b['rest_type'].value_counts(), color='r')
plt.title('Bar plot for the Restaurants in Bangalore')
plt.ylabel('amounts of order')
plt.xlabel('Restaurants type')
```

```python
#Let's try zooming in if we are going to see the restuarant types
#using the figure()

#the zooming method
plt.figure(figsize=(20,12))      #the zoming function must be at the top of alll the code

plt.bar(x=b['rest_type'].value_counts().index,
       height=b['rest_type'].value_counts(), color='r')
plt.title('Bar plot for the Restaurants in Bangalore')
plt.ylabel('amounts of order')
plt.xlabel('Restaurants type')

```

```python
#since its still give us the same problem and overlapping,
#lets try using the rotaion method

#the zooming method
plt.figure(figsize=(20,12))

plt.bar(x=b['rest_type'].value_counts().index,
       height=b['rest_type'].value_counts(), color='r')
plt.title('Bar plot for the Restaurants in Bangalore')
plt.ylabel('amounts of order')
plt.xlabel('Restaurants type')


#Using the gcf()- means get the original size and autofmt()_xdate
#to rotate the x axis and align it
plt.gcf().autofmt_xdate()
```

```python
#We can also try to use the ticks() function


#the zooming method
plt.figure(figsize=(20,12))

plt.bar(x=b['rest_type'].value_counts().index,
       height=b['rest_type'].value_counts(), color='r')
plt.title('Bar plot for the Restaurants in Bangalore')
plt.ylabel('amounts of order')
plt.xlabel('Restaurants type')

#using the ticks() function to rotate the x-axis
plt.xticks(rotation='vertical')
```

```python
#Use the plt.show() function to get rid of those arrays listed before the plot


#the zooming method
plt.figure(figsize=(20,12))

plt.bar(x=b['rest_type'].value_counts().index,
       height=b['rest_type'].value_counts(), color='r')
plt.title('Bar plot for the Restaurants in Bangalore')
plt.ylabel('amounts of order')
plt.xlabel('Restaurants type')

#using the ticks() function to rotate the x-axis
plt.xticks(rotation='vertical')
plt.yticks(rotation=45)

plt.show()
```
