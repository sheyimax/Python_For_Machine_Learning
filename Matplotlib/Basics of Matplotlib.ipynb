### Generating a dictionary of an array with different values for a scatter plot

```python
np.random.seed(1236790)                 #make the array constant values

#let's generate a dictionary of differents array
data = {'a' : np.arange(50),                #generate 50 numbers(0-50)     
        'c' : np.random.randint(0,50,50),   #generate an integers of 0-50
        'd' : np.random.randn(50) }         #generate negatives values

data['b'] = data['a'] + 10 * np.random.rand(50)
data['d'] = np.abs(data['d']) * 50          #convert 'd' to positive values

#plotting the graph
fig, e = plt.subplots(figsize=(5.8,3))
e.scatter('a', 'b', c='c', s='d', data = data)
plt.title('A Scatter Plot')
plt.xlabel('Entry a')
plt.ylabel('Entry b')
```

### Plotting multiple sets of data by passing in multiple sets of arguments of X and Y axis in the plot() method

```python
e = np.arange(5)
f = e**5
h = np.arange(2,10,3)
i = h-4

figure, g = plt.subplots(figsize=(5,3))
g.plot([1,2,3,4],[40,10,30,20],'go', e, f, 'r^', i,h, 'm*')
plt.title('A Sales Plot')
plt.xlabel('Month')
plt.ylabel('Sales')
```

 ### Sine wave plot

### Bar graph

```python
month = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May'])
sales = np.array([40,30,60,85,70])

figure, sales_month = plt.subplots(figsize=(8,6))
#sales_month.bar(month, sales, color='m')
sales_month.bar(height=sales, x=month,  color='m')
plt.title('Bar Plot for Sales/month')
plt.xlabel('Months')

plt.ylabel('Sales')
```

```python
#Changing it to horizontal using the barh() syntax


month = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May'])
sales = np.array([40,30,60,85,70])

figure, sales_month = plt.subplots(figsize=(8,6))
#sales_month.barh(month, sales, color='m')
sales_month.barh(width=sales, y=month,  color='m')
plt.title('Bar Plot for Sales/month')
plt.xlabel('Months')
plt.ylabel('Sales')
```

```python
### Use of Legend in Bar Plot
```

```python
month = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May'])
LAGsales = np.array([40,30,60,85,70])
ABJsales = np.array([25,48,72,38,54])
index = np.arange(1,6)
width = 0.40

figure, sales_month = plt.subplots(figsize=(8,6))
sales_month.bar(index, LAGsales, width, color='m', label='Lag Sales')
sales_month.bar(index+width, ABJsales, width, color='g', label='Abj Sales')
plt.title('Horizontally Stacked Bar Graph')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(index+width/2, month)
plt.legend(loc='best')
plt.show()
```

### More On Pie Chart

```python
month = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May'])
LAGsales = np.array([40,30,60,85,70])

figure, sales_month = plt.subplots(figsize=(8,6))
sales_month.pie(labels=month, x=LAGsales, shadow=True, 
                startangle=45, explode=[0.05,0.1,0.05,0.1,0])
plt.title('Sales Pie Chart')
plt.legend(loc='upper left')
plt.show()
```

### Multiple plots in one single graph

```python
#lets create a graph of 2x2 plots with graph on each plot
x = np.linspace(0, 2* np.pi, 100)
y = np.sin(x**2)

figure, axes = plt.subplots(2,2)

#for upper-left side
axes[0,0].plot(x, y, 'r')
axes[0,0].set_title('1st Subplot')

#for upper-right side
axes[0,1].plot(x,y, 'm')
axes[0,1].set_title('2nd Subplot')

#for lower-left side
axes[1,0].plot(x,y, 'g')
axes[1,0].set_title('3rd Subplot')

#for lower-right side
axes[1,1].plot(x,y, 'b')
axes[1,1].set_title('4th Subplot')

# To hide x-labels and tick-labels for 1st and 2nd plot 
for ax in axes.flat:
    ax.label_outer()

#general name
plt.suptitle('Sinusoidal Wave')
plt.show()
```
